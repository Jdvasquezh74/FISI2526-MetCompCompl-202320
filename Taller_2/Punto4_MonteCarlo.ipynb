4.1

El valor analítico de la integral es $\frac{1}{2}(1+e^{-\pi})$

```python
import numpy as np
import matplotlib.pyplot as plt

realvalue = (1+np.exp(-np.pi))/2

def func_integrate(x):
    return np.exp(-x)*np.sin(x)

muestras = np.linspace(10000,10000000,100,dtype="int")

a = 0
b = np.pi

cota = 1/np.sqrt(muestras)

error = []

for i in muestras:
    x = np.random.uniform(a,b,i)
    fi = func_integrate(x)
    I = (b-a)*sum(fi)/i
    err = (np.abs(realvalue-I)/realvalue)
    error.append(err)

plt.figure(dpi=100)
plt.plot(muestras, error)
plt.plot(muestras, cota,'--')
plt.legend(["Porcentaje de error", "Cota Máxima"])
plt.title("Porcentaje de error con respecto al número de muestras")
plt.xlabel("Número de muestras N")
plt.ylabel("Porcentaje de error %")
plt.show()


```

Haciendo el cambio de coordenadas de la integral a esféricas:

$\int_{0}^{\pi}\int_{0}^{2\pi}\int_{0}^{1} r^2sin(\phi)sin(r^2)e^{r^2}drd\theta d\phi$

```python
N = 10000000

theta = np.random.uniform(0,2*np.pi,N)
phi = np.random.uniform(0,np.pi,N)
r = np.random.uniform(0,1,N)

func = lambda r, phi: r**2*np.sin(phi)*np.sin(r**2)*np.exp(r**2) 
f = func(r,phi)

I = sum(f)*2*np.pi*np.pi/N
print(I)
```
